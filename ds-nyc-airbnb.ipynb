## Separating numeric and categorical variables and explore this

```python
numeric_data = df.select_dtypes(include=[np.number])
cat_data = df.select_dtypes(exclude=[np.number])
print(numeric_data.shape, cat_data.shape)
```

### Numeric data

```python
numeric_data.columns
```

```python
num_data = numeric_data.drop(columns=['id'])
num_data.head()
```

```python
corr = num_data.corr()
sns.heatmap(corr)
```

```python
corr['price'].sort_values(ascending=False)
```

```python
import urllib

plt.figure(figsize=(10,8))

i=urllib.request.urlopen('https://upload.wikimedia.org/wikipedia/commons/e/ec/Neighbourhoods_New_York_City_Map.PNG')
nyc_img=plt.imread(i)

long_max = df['longitude'].max()
long_min = df['longitude'].min()
lat_max = df['latitude'].max()
lat_min = df['latitude'].min()
plt.imshow(nyc_img,zorder=0,extent=[long_min, long_max, lat_min, lat_max])
ax=plt.gca()

price_500 = num_data[num_data['price'] < 500]
price_500.plot(kind='scatter', x='longitude', y='latitude', label='availability_365', c='price', ax=ax, 
           cmap=plt.get_cmap('jet'), colorbar=True, alpha=0.4, zorder=5)
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

_data = df[['price', 'neighbourhood_group', 'room_type']]
model = smf.ols(formula='price ~ C(neighbourhood_group) + C(room_type)', data=_data).fit()
print(model.summary())
```

```python
table = sm.stats.anova_lm(model, typ=3)
print(table)
```
